# Data Distribution

```python
#Plot histograms for numerical columns to visualize their distribution
df1.hist(bins = 30, figsize=(15,10))
plt.show()
```

```python
#plot a boxplot to look for outliers in numerical columns
columns = ['Total Avg Rating', 'Number of Pages', 'Kindle Price']
df1.boxplot(column = columns, figsize=(15,10))
plt.show()
```

# Correlation Analysis

```python
#To understand the relationships between numerical variables
data = df1.corr()
data
```

```python
#correlation between variables
plt.figure(figsize= (15,10))
sns.heatmap(data, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.show()
```

# EDA on Category

```python
#Grouped Analysis
grouped = df1.groupby('Readers Choice Category').mean()
grouped
```

# Time Series Analysis

```python
#Date Parsing
df1['First Published date'] = pd.to_datetime(df1['First Published date'])
df1['month'] = df1['First Published date'].dt.month
```

```python
#Analyze trends over time
df1.groupby('month')['Total Avg Rating'].mean().plot(kind='line')
plt.show()
```

# PRICE ANALYSIS

```python
#Analyze the distribtuion of book prices
df1['Kindle Price'].plot(kind = 'hist', bins=30)
plt.show()
```

```python
#Text Data Analysis
from wordcloud import WordCloud

text = ' '.join(df1['Book Description'].dropna().tolist())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
